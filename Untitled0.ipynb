```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("/content/Document from venky")
print(df.head())
```

```python
print(df.info())
```

```python
print(df.columns)
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
print(df['City'].unique())
```

```python
print(df['City'].value_counts())
```

```python
df['City'] = df['City'].str.title()
```

```python
print(df.duplicated().sum())
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
print(df['Age'].isnull().sum())
```

```python
df['Department']=df['Department'].fillna(df['Department'].mode()[0])
```

```python
print(df['Department'].isnull().sum())
```

```python
df['Attendance (%)']=df['Attendance (%)'].fillna(df['Attendance (%)'].mean())
```

```python
print(df['Attendance (%)'].isnull().sum())
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
print(df['Gender'].isnull().sum())
```

```python
print(df.describe())
```

```python
correlation1= df['Attendance (%)'].corr(df['Marks'])
print(correlation1)
```

```python
def attendance_category(x):
    if x >= 90:
        return 'Excellent'
    elif x >= 75:
        return 'Good'
    else:
        return 'Poor'

df['Attendance_Category'] = df['Attendance (%)'].apply(attendance_category)
```

```python
print(df.head())
```

```python
df['Attendance_Status'] = df['Attendance (%)'].apply(
    lambda x: 'Regular' if x >= 75 else 'Irregular'
)
```

```python

sns.countplot(x='Department', data=df)
plt.title('Students by Department')
plt.show()
```

```python
gender_count = df['Gender'].value_counts()

plt.pie(gender_count,
        labels=gender_count.index,
        autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
sns.scatterplot(x='Attendance (%)',
                y='Marks',
                data=df)
plt.show()
```

```python
sns.countplot(y='City', data=df)
plt.show()
```
